```python
df=pd.read_csv("student Exam Scores.csv")
print(df.head())
```

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head)
```

```python
#change weekly study hours column
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

```python
#gender distribution
plt.figure(figsize=(4,4))
ax=sns.countplot(data=df,x="Gender")
plt.title("GENDER DISTRIBUTION")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#from the above chart we have analysed that the number of female in the data is more than that of males.
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.show
```

```python
#from the above chart we have concluded that parent's education has a impact in their children
```

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(data=df,x="ParentMaritalStatus")
plt.title("Marital Status")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
sns.heatmap(gb1,annot=True)
plt.show
```

```python
'''from the above analysis we can state that the marital status of a parent
does not have any kind of negative impact on the students.#
```

```python
plt.figure(figsize=(4,4))
ap=sns.countplot(data=df,x="WklyStudyHours")
plt.title("STudy HOurs")
ap.bar_label(ap.containers[0])
plt.show()
```

```python
gb2=df.groupby("WklyStudyHours").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb2)
```

```python
sns.heatmap(gb2,annot=True)
plt.show
```

```python
#from the above analysi we can say that student who practices math for less than 5hr they are getting
#a bit of less marks than that of the students who are practicing math more  tahn 10hr
#other than that students with different study house doesn't have so much difference in their result.
```

```python
plt.figure(figsize=(4,4))
a=sns.countplot(data=df,x="TestPrep")
plt.title("Test Preperation")
a.bar_label(a.containers[0])
plt.show()
```

```python
gb3=df.groupby("TestPrep").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb3)
```

```python
sns.heatmap(gb3,annot=True)
plt.show
```

```python
#from the above analysis we can say that probably very less amount of students have completed their test preparation ,
#and certaily they got far better marks in all the three aspects,from the students who didn't completed test preparation.
```

# distribution of ethnic groups

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()
l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of groups")
plt.show()
```
